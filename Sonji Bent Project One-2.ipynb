# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Sum the values for each school
sum_values = selected_rows.groupby('School Name')['Number Scoring 65 or Above'].sum()

# Calculate the total count of students for each school
total_count = selected_rows.groupby('School Name')['Total Tested'].sum()

# Calculate the percentage of scores above 65% for each school
percentage_above_65 = (sum_values / total_count) * 100

# Create a bar graph of the percentages
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust the figure size
percentage_below_65.plot(kind='bar', ax=ax, color=['#FF6F61', '#6B5B95'])  # Customize the bar color

ax.set_xlabel('School Name', fontsize=12)  # Set the x-axis label and font size
ax.set_ylabel('Percentage Above 65%', fontsize=12)  # Set the y-axis label and font size
ax.set_title('Percentage of Scores Above 65%\nU.S. History and Government', fontsize=14, fontweight='bold')
```

Two 6 - 12 schools in District 17 in Brookly were compared.  The U.S. History and Government regents was compared.  The barplot shows that, Medgar Evers College Preparatory School had a lower percentage rate of students with a Percentage of Scores Above 65% in the U.S. History and Government Regents. Science, Technology and Research Early College High School had twice as many students with a Percentage of Scores Above 65%.
