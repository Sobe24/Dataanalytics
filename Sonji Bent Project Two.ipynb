# Data Cleaning

```python
# change values in a column
df['race'] = df['race'].replace('?','Other')
df['weight'] = df['weight'].replace('?','0')
print(df)
```

```python
# Replace multiple values in one column
df.replace({'time_in_hospital' : { '1' : '1 Day', '2' : '2 Days', '3' : '3 Days', '4' : '4 Days', '5' : '5 Days', '6' : '6 Days', '7' : '7 Days', '8' : '8 Days', '9' : '9 Days', '10' : '10 Days', '11' : '11 Days', '12' : '12 Days', '13' : '13 Days', '14' : '14 Days'}})
```

```python
#set a new index
df=df.set_index('encounter_id')
df.head()
```

```python
df.head(20)
```

```python
# Convert to numeric
cols = ['diag_1', 'diag_2', 'diag_3']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

# Analyze Data

```python
df.describe()
```

```python
df2 = df[['patient_nbr','race','gender','age','weight','time_in_hospital','A1Cresult','change','diabetesMed','readmitted']]
```

```python
df3_genderm = df2[df2['gender'].str.contains('Male')]
grouped_data = df3_genderm.groupby('race')
genderm_counts = grouped_data['time_in_hospital'].value_counts()
total_rows_in_each_genderm_group = grouped_data.size()
genderm_percentage = genderm_counts / total_rows_in_each_genderm_group * 100
```

```python
df3_genderf = df2[df2['gender'].str.contains('Female')]
grouped_data = df3_genderf.groupby('race')
genderf_counts = grouped_data['time_in_hospital'].value_counts()
total_rows_in_each_gender_group = grouped_data.size()
genderf_percentage = genderf_counts / total_rows_in_each_genderf_group * 100
```
